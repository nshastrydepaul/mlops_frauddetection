```python
import pandas as pd

chunks = pd.read_csv("../data/raw/data.csv", chunksize=200_000)

dfs = []
for chunk in chunks:
    # sample proportionally from each chunk
    dfs.append(chunk.sample(frac=0.0125, random_state=42))  # ~100K overall

df_sample = pd.concat(dfs).sample(n=100_000, random_state=42)
df_sample.to_csv("../data/raw/data_100k.csv", index=False)

print("Saved data_100k.csv:", df_sample.shape)
```
